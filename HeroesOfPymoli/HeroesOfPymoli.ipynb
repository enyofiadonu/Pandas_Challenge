```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head(10)
```

```python
#Finding number of unique players

player_demographics = purchase_data.loc[:,['Gender','SN','Age']]
player_demographics = player_demographics.drop_duplicates()
tot_players=player_demographics.count()[0]
tot_players_df = pd.DataFrame({'Total Players': [tot_players]})
tot_players_df
```

```python
#Purchasing Analysis
#Finding numver of unique items, their average price, number of purchases,and total revenue

unique_item_count= len(purchase_data['Item ID'].unique())
avg_price = round(float(purchase_data['Price'].mean()),2)
purchases_count = len(purchase_data['Price'])
price_sum = float(purchase_data['Price'].sum())

item_df = pd.DataFrame({
    'Number of Unique Items': [unique_item_count],
    'Average Price': '$'+str(avg_price),
    'Number of Purchases': [purchases_count],
    'Total Revenue': '$'+str(price_sum)
})
item_df
```

```python
#Gender Demographics
#Find the percentage and count of all of the genders of the players 

gender_df = purchase_data[['Gender','SN']].drop_duplicates(subset = 'SN')
gender_count = gender_df['Gender'].value_counts()
gender_total = gender_count.sum()
gender_percent = round((gender_count/gender_total)*100,2)

gender_count_df = pd.DataFrame(gender_count)
gender_percent_df = pd.DataFrame(gender_percent)

gender_demographics_df = gender_count_df.merge(gender_percent_df,left_index = True, right_index = True)
gender_demographics_df.columns=['Total Count','Percentage of Players']
gender_demographics_df['Total Count'] = gender_demographics_df['Total Count'].map('${:,.2f}'.format)
gender_demographics_df['Percentage of Players'] = gender_demographics_df['Percentage of Players'].map('{:,.2f}%'.format)
gender_demographics_df
```

```python
#Purchasing Analysis by Gender
#Find the purchase count, avg. purchase price, purchase total per person by gender

gender_df2 = purchase_data.groupby('Gender')

purchase_count = round(gender_df2['Purchase ID'].count(),0)
avg_purchase_price = round(gender_df2['Price'].mean(),2)
total_value = round(gender_df2['Price'].sum(),2)
purchase_gender = round(total_value/gender_count,2)

analysis_df = pd.DataFrame([purchase_count,avg_purchase_price,total_value,purchase_gender])
purchasing_analysis_gender_df = analysis_df.T
purchasing_analysis_gender_df.columns = ['Purchase Count','Average Purchase Price','Total Purchase Value','Average Total Purchase per Gender']
purchasing_analysis_gender_df['Average Purchase Price'] = purchasing_analysis_gender_df['Average Purchase Price'].map('${:,.2f}'.format)
purchasing_analysis_gender_df['Total Purchase Value'] = purchasing_analysis_gender_df['Total Purchase Value'].map('${:,.2f}'.format)
purchasing_analysis_gender_df['Average Total Purchase per Gender'] = purchasing_analysis_gender_df['Average Total Purchase per Gender'].map('${:,.2f}'.format)
purchasing_analysis_gender_df
```

```python
#Age Demographics
#Find the total count and what percent of the total players by age group
bins = [0,9,14,19,24,29,34,39,100]
bin_labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
purchase_data['Age Group'] = pd.cut(purchase_data['Age'],bins,labels=bin_labels)
unique_age_df = purchase_data[['Age Group','SN']].drop_duplicates(subset='SN')

age_df = unique_age_df.groupby('Age Group').count()
age_counts = age_df['SN']
age_percentages = round(age_counts/tot_players*100,2)

age_df2 = pd.DataFrame([age_counts,age_percentages])
age_demographics_df = age_df2.T
age_demographics_df.columns = ['Total Count','Percentage of Players']
age_demographics_df['Percentage of Players'] = age_demographics_df['Percentage of Players'].map('{:,.2f}%'.format)
age_demographics_df
```

```python
#Purchasing Analysis by Age
#Find the purchase count, avg. purchase price, purchase total per person by age group
#the bins were created in the previous cell

reduced_rows = purchase_data[['Age Group','SN','Price']]
reduced_groupby = reduced_rows.groupby('Age Group')
purchase_count = reduced_groupby['Price'].count()
avg_price = round(reduced_groupby['Price'].mean(),2)
total_spent = round(reduced_groupby['Price'].sum(),2)
spending_per_person = round(total_spent/age_counts,2)

analysis_age = pd.DataFrame([purchase_count,avg_price,total_spent,spending_per_person])
purchasing_analysis_age_df = analysis_age.T
purchasing_analysis_age_df.columns = ['Purchase Count', 'Average Purchase Price', 'Total Purchase Value','Avg Total Purchase per Person']
purchasing_analysis_age_df['Average Purchase Price'] = purchasing_analysis_age_df['Average Purchase Price'].map('${:,.2f}'.format)
purchasing_analysis_age_df['Total Purchase Value'] = purchasing_analysis_age_df['Total Purchase Value'].map('${:,.2f}'.format)
purchasing_analysis_age_df['Avg Total Purchase per Person'] = purchasing_analysis_age_df['Avg Total Purchase per Person'].map('${:,.2f}'.format)
purchasing_analysis_age_df
```

```python
#Top Spender
#sort through data frame to find the player that spent the most in the game

SN_df = purchase_data.groupby('SN')
purchase_count = SN_df['Gender'].count()
avg_spending = round(SN_df['Price'].mean(),2)
tot_price = round(SN_df['Price'].sum(),2)

analysis_price = pd.DataFrame([purchase_count,avg_spending,tot_price])
top_spender_analysis_df = analysis_price.T
top_spender_analysis_df.columns = ['Purchase Count','Average Purchase Price','Total Purchase Value']
top_spender_analysis_df = top_spender_analysis_df.sort_values('Total Purchase Value',ascending=False).reset_index()
top_spender_analysis_df['Average Purchase Price'] = top_spender_analysis_df['Average Purchase Price'].map('${:,.2f}'.format)
top_spender_analysis_df['Total Purchase Value'] = top_spender_analysis_df['Total Purchase Value'].map('${:,.2f}'.format)
top_spender_analysis_df.head()
```

```python
#Most Popular Items
#Sort through the data to find the most popular items purchased

items_df = purchase_data.groupby(['Item ID','Item Name'])
purchase_count = items_df['Gender'].count()
avg_spending = round(items_df['Price'].mean(),2)
tot_purchase = round(items_df['Price'].sum(),2)

analysis_items = pd.DataFrame([purchase_count,avg_spending,tot_purchase])
most_popular_items_df = analysis_items.T
place_holder = analysis_items.T
most_popular_items_df.columns = ['Purchase Count','Item Price','Total Purchase Value']
most_popular_items_df = most_popular_items_df.sort_values('Purchase Count',ascending=False)
most_popular_items_df['Item Price'] = most_popular_items_df['Item Price'].map('${:,.2f}'.format)
most_popular_items_df['Total Purchase Value'] = most_popular_items_df['Total Purchase Value'].map('${:,.2f}'.format)
most_popular_items_df.head()
```

```python
#Most Profitable Items
#sort throught the data to find the item with the maximum total purchase value

place_holder.columns = ['Purchase Count','Item Price','Total Purchase Value']
most_profitable_df = place_holder.sort_values('Total Purchase Value', ascending=False)
most_profitable_df['Item Price'] = most_profitable_df['Item Price'].map('$ {:,.2f}'.format)
most_profitable_df['Total Purchase Value'] = most_profitable_df['Total Purchase Value'].map('${:,.2f}'.format)
most_profitable_df.head()
```
